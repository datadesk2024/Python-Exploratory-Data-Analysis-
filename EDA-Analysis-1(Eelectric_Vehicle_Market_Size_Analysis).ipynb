```python
#Lets Analyze EV Adoption Over Time by Visualizig  the number of  EV registered Vehicles by Model Year

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")


plt.figure(figsize=(12, 6))
ev_adoption_by_year = ev_data['Model Year'].value_counts().sort_index()
sns.barplot(x=ev_adoption_by_year.index, y=ev_adoption_by_year.values, palette='viridis')
plt.title('Electric Vehicle Adoption Over Time')
plt.xlabel('Model Year')
plt.ylabel('Number of Vehicles Registered')
plt.xticks(rotation=45)
plt.show()

```

```python
# Lets start by selecting the top 3 Countries based on EV Registrations

ev_country_distribution = ev_data['County'].value_counts()
top_countries = ev_country_distribution.head(3).index

#filtering the dataset for these top countries
top_ev_data = ev_data[ev_data['County'].isin(top_countries)]

ev_city_distribution_top_countries = top_ev_data.groupby(['County', 'City']).size().sort_values(ascending=False).reset_index(name='Number of Vehicles')
print(ev_city_distribution_top_countries)


top_cities = ev_city_distribution_top_countries.head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='Number of Vehicles', y='City', hue='County', data=top_cities, palette="magma")
plt.title('Top 10 Cities with the Most Electric Vehicles Registered')
plt.xlabel('Number of Vehicles Registered')
plt.ylabel('City')
plt.xticks(rotation=45)
plt.legend(title='County')
plt.show()
```

```python
# Analyzing the distribution of Electric Vehciles types

ev_type_distribution = ev_data['Electric Vehicle Type'].value_counts()
print(ev_type_distribution)

plt.figure(figsize=(10, 5))
sns.barplot(x=ev_type_distribution.index, y=ev_type_distribution.values, palette='rocket')
plt.title('Distribution of Electric Vehicle Types')
plt.xlabel('Number of Vehicles Registered')
plt.ylabel('Electric Vehicle Types')
plt.xticks(rotation=45)
plt.show()

```

```python


#Analyzing the most popular manufactureres

ev_manufacturer_distribution = ev_data['Make'].value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=ev_manufacturer_distribution.index[:10], y=ev_manufacturer_distribution.values[:10], palette='cubehelix')
plt.title('Top 10 Manufacturers with the Most Electric Vehicles Registered')
plt.xlabel('Manufacturer')
plt.ylabel('Number of Vehicles Registered')
plt.xticks(rotation=45)
plt.show()


```

```python
# Selecting the top3 Manufacturers based on Number of Vehciles Types Regsitered

top_3_makes = ev_manufacturer_distribution.head(3).index

top_ev_data = ev_data[ev_data['Make'].isin(top_3_makes)]

ev_model_distribution_top_makes = top_ev_data.groupby(['Make', 'Model']).size().sort_values(ascending=False).reset_index(name='Number of Vehicles')
print(ev_model_distribution_top_makes)



top_models = ev_model_distribution_top_makes.head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='Number of Vehicles', y='Model', hue='Make', data=top_models, palette="rocket")
plt.title('Top 10 Models with the Most Electric Vehicles Registered')
plt.xlabel('Number of Vehciles Registered')
plt.ylabel('Model')
plt.xticks(rotation=45)
plt.legend(title='Make')
plt.show()
```

```python

# We will calculate the CAGR between a recent year with complete data and the first year

from  scipy.optimize import curve_fit

import numpy as np

filtered_year = ev_data[ev_data['Model Year'] == 2023]
filtered_year = filtered_year.dropna()

def exp_growth(t, a, b):
    return a * np.exp(b * t)

x_data = np.arange(len(filtered_year))
y_data = filtered_year.values

params, covariance  = curve_fit(exp_growth, x_data, y_data)

forecast_years = np.arange(len(filtered_year), len(filtered_year) + 5)
forecast_values = exp_growth(forecast_years, *params)

forecasted_evs = pd.Series(forecast_values, index=forecast_years)

print(forecasted_evs)
```
